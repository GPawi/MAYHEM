# Check on Proxy Resolution

## Importing needed libraries 

```python
%matplotlib inline
%config IPCompleter.greedy=True
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlalchemy
password = open('../password.txt', 'r').read()
```

## Organics

#### Connecting to the database

```python
engine = sqlalchemy.create_engine("postgresql://postgres:%s@localhost/MAYHEM" % password)
con = engine.connect()
```

#### Insert age depth model output and organic data into dataframe

```python
df_model_output = pd.read_sql('modeloutput', con)
df_model_output = df_model_output.sort_values(by = 'measurementid', ignore_index = True)
df_organics = pd.read_sql('organic', con)
```

#### Close the connection to the database

```python
con.close()
```

## TOC
Note:  TN is measured together with TOC and will be later added as proxy for this resolution

#### Merge TOC data with age depth model median output

```python
df_toc_density = pd.merge(df_organics[['measurementid','toc']], 
                              df_model_output[['measurementid', 'modeloutput_median']],
                             how = 'outer',
                             on = ['measurementid'])
df_toc_density['coreid'],df_toc_density['compositedepth'] = df_toc_density['measurementid'].str.split(' ', 1).str
df_toc_density['compositedepth'] = df_toc_density['compositedepth'].astype(dtype = float)
```

**Linear interpolate age depth model for TOC values that do not have an age assigned to their depth\
Further calculate the proxy resolution as difference between each proxy point**

```python
df_toc_resolution = pd.DataFrame(columns = ['coreid', 'compositedepth', 'modeloutput_median', 'proxyresolution'])

for i in range(0, len(df_toc_density['coreid'].unique())):
    df_toc_density_slice = df_toc_density[(df_toc_density['coreid'] == df_toc_density['coreid'].unique()[i])]
    df_toc_density_slice = df_toc_density_slice.sort_values(by = 'compositedepth', ignore_index = True)
    df_toc_density_slice['modeloutput_median'] = df_toc_density_slice['modeloutput_median'].interpolate()
    df_toc_density_slice = df_toc_density_slice.dropna()
    for j in range(0, len(df_toc_density_slice)-1):
        individual_result = pd.DataFrame(np.array([[df_toc_density_slice.iloc[j+1,3],
                                                df_toc_density_slice.iloc[j+1,4],
                                                df_toc_density_slice.iloc[j+1,2],
                                                round(((df_toc_density_slice.iloc[j+1,2] - df_toc_density_slice.iloc[j,2])),6)]]), 
                                                columns = ['coreid','compositedepth', 'modeloutput_median', 'proxyresolution'])
        df_toc_resolution = df_toc_resolution.append(individual_result)

df_toc_resolution = df_toc_resolution.reset_index(drop = True)
df_toc_resolution = df_toc_resolution.astype(dtype = {'coreid' : str, 
                                                      'compositedepth' : float,
                                                      'modeloutput_median' : float,
                                                      'proxyresolution' : float})
```

#### Plotting the proxy resolution for TOC vs model output median

```python
### Changing the look of the graphic
SMALLER_SIZE = 12
MEDIUM_SIZE = 15
BIGGER_SIZE = 18
plt.rcdefaults()
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALLER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALLER_SIZE)    # fontsize of the tick labels
#####
plt.figure(figsize=(10,5))
ax = sns.scatterplot(x = 'modeloutput_median', y = 'proxyresolution', data = df_toc_resolution, hue='coreid', legend = 'brief')
plt.xlabel('cal yr BP (Median)', weight = 'bold')
plt.ylabel('Proxy resoultion for TOC data \n [yrs per proxy determination]', weight = 'bold')
plt.legend(ncol = 1, fontsize = 8, loc = 'right', bbox_to_anchor = (1.2, 0.5))
plt.savefig('Figures/TOC_TN_Proxy_Resolution.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

```python
df_model_output = pd.read_sql('modeloutput', con)
df_model_output = df_model_output.sort_values(by = 'measurementid', ignore_index = True)
df_grainsize = pd.read_sql('grainsize', con)
df_grainsize = pd.DataFrame(df_grainsize.groupby(by = 'measurementid'), columns = ['measurementid', 'other'])
```

#### Close the connection to the database

```python
con.close()
```

#### Merge grain size data with age depth model median output

```python
df_grainsize_density = pd.merge(df_grainsize[['measurementid', 'other']], 
                         df_model_output[['measurementid', 'modeloutput_median']],
                         how = 'outer',
                         on = ['measurementid'])
df_grainsize_density['coreid'],df_grainsize_density['compositedepth'] = df_grainsize_density['measurementid'].str.split(' ', 1).str
df_grainsize_density['compositedepth'] = df_grainsize_density['compositedepth'].astype(dtype = float)
```

**Linear interpolate age depth model for grain size values that do not have an age assigned to their depth\
Further calculate the proxy resolution as difference between each proxy point**

```python
df_grainsize_resolution = pd.DataFrame(columns = ['coreid', 'compositedepth', 'modeloutput_median', 'proxyresolution'])

for i in range(0, len(df_grainsize_density['coreid'].unique())):
    df_grainsize_density_slice = df_grainsize_density[(df_grainsize_density['coreid'] == df_grainsize_density['coreid'].unique()[i])]
    df_grainsize_density_slice = df_grainsize_density_slice.sort_values(by = 'compositedepth', ignore_index = True)
    df_grainsize_density_slice['modeloutput_median'] = df_grainsize_density_slice['modeloutput_median'].interpolate()
    df_grainsize_density_slice = df_grainsize_density_slice.dropna()
    for j in range(0, len(df_grainsize_density_slice)-1):
        individual_result = pd.DataFrame(np.array([[df_grainsize_density_slice.iloc[j+1,3],
                                                df_grainsize_density_slice.iloc[j+1,4],
                                                df_grainsize_density_slice.iloc[j+1,2],
                                                round(((df_grainsize_density_slice.iloc[j+1,2] - df_grainsize_density_slice.iloc[j,2])),6)]]), 
                                                columns = ['coreid','compositedepth', 'modeloutput_median', 'proxyresolution'])
        df_grainsize_resolution = df_grainsize_resolution.append(individual_result)

df_grainsize_resolution = df_grainsize_resolution.reset_index(drop = True)
df_grainsize_resolution = df_grainsize_resolution.astype(dtype = {'coreid' : str, 
                                                      'compositedepth' : float,
                                                      'modeloutput_median' : float,
                                                      'proxyresolution' : float})
```

#### Plotting the proxy resolution for grain size vs model output median

```python
### Changing the look of the graphic
SMALLER_SIZE = 12
MEDIUM_SIZE = 15
BIGGER_SIZE = 18
plt.rcdefaults()
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALLER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALLER_SIZE)    # fontsize of the tick labels
#####
plt.figure(figsize=(10,5))
ax = sns.scatterplot(x = 'modeloutput_median', y = 'proxyresolution', data = df_grainsize_resolution, hue='coreid', legend = 'brief')
plt.xlabel('cal yr BP (Median)', weight = 'bold')
plt.ylabel('Proxy resoultion for grain size data \n [yrs per proxy determination]', weight = 'bold')
plt.legend(ncol = 1, fontsize = 8, loc = 'right', bbox_to_anchor = (1.15, 0.5))
plt.savefig('Figures/Grain_Size_Proxy_Resolution.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

# Comparison of all proxies

#### Rename the columns of each resolution dataframe to the appropriate proxy

#### Combine all proxies in one dataframe

```python
df_comparison = pd.DataFrame(columns = ['coreid', 'compositedepth', 'modeloutput_median', 'proxyresolution', 'proxy'])
df_comparison = df_comparison.append([df_toc_resolution,
                                     df_element_resolution,
                                      df_d13c_resolution, 
                                      df_diatom_resolution,
                                      df_tc_resolution,
                                     df_pollen_resolution,
                                     df_grainsize_resolution,
                                     df_mineral_resolution, 
                                      df_chironomid_resolution])
```

```python
### Changing the look of the graphic
SMALLER_SIZE = 12
MEDIUM_SIZE = 15
BIGGER_SIZE = 18
plt.rcdefaults()
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALLER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALLER_SIZE)    # fontsize of the tick labels
#####
plt.figure(figsize=(10,5))
ax = sns.scatterplot(x = 'modeloutput_median', y = 'proxyresolution', data = df_comparison, hue='proxy', legend = 'brief', s = 30)
plt.ylim(0,10000)
plt.xlabel('Age [cal yr BP (Median)]', weight = 'bold')
plt.ylabel('Proxy resolution \n [yrs per proxy determination]', weight = 'bold')
plt.legend(ncol = 1, fontsize = 8, loc = 'right', bbox_to_anchor = (1.2, 0.5))
plt.savefig('Figures/Proxy_Comparison_One_Plot.png', dpi = 300)
plt.show()
```
