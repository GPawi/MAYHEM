# Age Control Plotting

## Importing needed libraries 

```python
%matplotlib inline
%config IPCompleter.greedy=True
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Preparing for stack bar plot with age determination

#### Using calibrated ages from extra Excel sheet in dataframe

```python
df_age = pd.read_excel('../COREDATA/AgeDetermination.xlsx', 'Data')
```

#### Pivot dataframe, so that every column belongs to individual CoreID

```python
df_age_pivot = df_age.pivot(columns = 'CoreID', values = 'yrBP')
df_age_pivot = df_age_pivot.apply(lambda x: pd.Series(x.dropna().values))
```

#### Get the numbers of samples per CoreID

```python
df_age_pivot.count()
```

#### Stacked bar plot of ages until 50,000 uncal years BP

```python
### Changing the look of the graphic
MEDIUM_SIZE = 15
BIGGER_SIZE = 18
plt.rcdefaults()
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rcParams["xtick.labeltop"] = False
plt.rcParams["xtick.top"] = False
plt.rcParams["xtick.labelbottom"] = False
plt.rcParams["xtick.bottom"] = True
#####
### Setting parameters for the plot
bins = np.arange(-1000,51000,1000)
colors = (sns.color_palette("Greens_r", 15) + sns.color_palette("Blues", 15) + sns.color_palette("Purples_r", 14) + sns.color_palette("RdPu", 14))
IDs = df_age.CoreID.unique()
#####
fig, ax = plt.subplots(figsize=(20,5))
plt.hist([df_age.loc[df_age.CoreID == x, 'yrBP'] for x in IDs], bins = bins, stacked=True, label=IDs, color = colors)
plt.xlim(-2000, 50000)
plt.yticks(np.arange(0,90,10))
plt.ylabel('Number of dating samples', weight = 'bold')
plt.savefig('Figures/age_distribution.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

## Preparing barplot for each interval

#### Arrays with the bins in 1000 years steps

```python
spacing_bins = np.arange(-1000,220000,1000) 
spacing_bins_without = np.delete(spacing_bins, -1)
```

#### Stack dataframe back to a dataframe sorted by ages 

```python
df_age_pivot = df_age_pivot.stack().reset_index()
```

#### Bin each age into the appropriate age bin

```python
df_age_pivot['binned'] = pd.cut(df_age_pivot[0], spacing_bins, right = False, labels = spacing_bins_without)
```

#### Count the number of ages per bin per CoreID and drop CoreIDs that are not in the age bin

```python
counts = df_age_pivot.groupby(['binned','CoreID']).count()
counts = counts.drop(columns = 'level_0').rename(columns = {0 : 'counts'})
na_counts = counts.dropna()
```

#### Create a Series for plotting 

```python
cores_per_interval = na_counts.reset_index().groupby('binned')['CoreID'].nunique()
```

#### Plotting the barplot

```python
### Changing the look of the graphic
MEDIUM_SIZE = 15
BIGGER_SIZE = 18
plt.rcdefaults()
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rcParams["xtick.labeltop"] = True
plt.rcParams["xtick.top"] = True
plt.rcParams["xtick.labelbottom"] = False
plt.rcParams["xtick.bottom"] = False
#####
### Setting parameters for the plot
bins = np.arange(-1000, 221000, 1000)
#####
plt.figure(figsize=(20,3))
plt.hist(cores_per_interval.keys(), weights = cores_per_interval, bins = bins)
plt.xlim(-2000,50001)
plt.ylim(0,35)
plt.xlabel('cal yr BP', weight = 'bold',labelpad = -190)
plt.ylabel('Numbers of cores \n per interval', weight = 'bold')
plt.savefig('Figures/histogram_number_of_cores_50000.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
